```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
# Parameters for the Ornstein-Uhlenbeck process
theta = 1
sigma = 0.5
initial_state = 0
t_span = (0, 20)
dt = 0.001

# Euler-Maruyama method for the Ornstein-Uhlenbeck process
def euler_maruyama_ou(x0, t_span, dt, theta, sigma):
    t0, tf = t_span
    t = np.linspace(t0, tf, int((tf - t0) / dt) + 1)
    x = np.zeros(len(t))

    x[0] = x0

    for i in range(len(t) - 1):
        noise_forcing = np.sqrt(dt) * np.random.normal(0, 1)
        x[i + 1] = x[i] - theta * x[i] * dt + sigma * noise_forcing

    return t, x
```

```python
# Generate the data for the Ornstein-Uhlenbeck process
t, x = euler_maruyama_ou(initial_state, t_span, dt, theta, sigma)

# Create a DataFrame with x values
#df = pd.DataFrame({'x': x})


# Save the DataFrame as a CSV file
#df.to_csv('./ou_process_data.csv', index=False)
```

```python
num_bins = 10  # You can adjust the number of bins to get better resolution
x_hist, x_bins = np.histogram(x, bins=num_bins, density=True)

plt.figure()
plt.plot(x_bins[:-1], x_hist, label='x PDF')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
# Equilibrium probability density function
x_eq = np.linspace(-3, 3, 100)
x_eq_pdf = np.sqrt(theta / (2 * np.pi * sigma ** 2)) * np.exp(-theta * x_eq ** 2 / (2 * sigma ** 2)) # PDF of the equilibrium distribution

plt.figure()
plt.plot(x_bins[:-1], x_hist, label='x PDF')
plt.plot(x_eq, x_eq_pdf, label='Equilibrium PDF')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
# Normalized ACF using FFT
def acf_fft(x):
    x = x - np.mean(x)
    n = len(x)
    f = np.fft.fft(x, n=2 * n)
    acf = np.real(np.fft.ifft(f * np.conjugate(f))[:n])
    acf = acf / acf[0]
    return acf

# Plot the ACF
plt.figure()
plt.plot(t, acf_fft(x), label='ACF')
plt.xlabel('Time')
plt.ylabel('ACF')
plt.legend()
plt.show()
```

```python
#Generate the data for the damped oscillation problem
def damped_oscillation(t, y, m, c, k):
    a = -c / m * y[1] - k / m * y[0]
    return [y[1], a]

if __name__ == "__main__":
    # Damped oscillation parameters
    m = 20.0
    c = 10.0
    k = 2e3

    # Initial conditions (displacement and velocity)
    y0 = [0, 0.1]

    # Time range
    t_start = 0
    t_end = 10
    t_points = 1000

    # Solve the damped oscillation problem
    t_eval = np.linspace(t_start, t_end, t_points)
    sol = solve_ivp(damped_oscillation, (t_start, t_end), y0, args=(m, c, k), t_eval=t_eval)

    # Save the data to a CSV file
    df = pd.DataFrame({"t": sol.t, "y": sol.y[0], "ydot": sol.y[1]})
    df.to_csv("./damped_data.csv", index=False)
```

```python
def generate_data():
    m = 20.0
    damping = 10.0
    dt = .001
    T = 1000

    t = np.linspace(0, T, int(T / dt) + 1)
    v = np.zeros(t.shape)

    v[0] = 0.1

    for i in range(0, len(t) - 1):
        v[i + 1] = v[i] - damping * v[i] * dt / m + np.sqrt(dt) * np.random.normal(0, 1)

    # Save the data to a CSV file
    df = pd.DataFrame({"t": t, "v": v})
    df.to_csv("./langevin_data.csv", index=False)

    plt.figure(figsize=(10, 6))
    plt.plot(t,v,'-')

generate_data()
```
