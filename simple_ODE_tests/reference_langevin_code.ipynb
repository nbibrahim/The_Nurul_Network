```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
epsilon = 1
r = 1
dt = .001
T = 1000

t = np.linspace(0,T,int(T/dt)+1)
y = np.zeros(t.shape)

y[0] = 0


for i in range(0,len(t)-1):
    y[i+1] = y[i]-r*y[i]*dt+np.sqrt(dt)*np.random.normal(0,1)*np.sqrt(epsilon)
```

```python
plt.plot(t,y,'-')
```
