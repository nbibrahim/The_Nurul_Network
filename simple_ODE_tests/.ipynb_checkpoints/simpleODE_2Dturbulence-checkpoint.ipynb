```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
k = 12 * np.pi
k_e_squared = 180 * np.pi ** 2
k_p_squared = 544 * np.pi ** 2
r_m = 0.1

W_e = np.array([[-1, 1 / (15 * np.pi)],
                [-1200 * np.pi, -1]])

W_p = np.array([[-1, -3 / (136 * np.pi)],
                [1200 * np.pi, -1]])

W = np.block([[W_e, np.zeros((2, 2))],
              [np.zeros((2, 2)), W_p]])

L_e = np.array([[-58 * np.pi / 5, 0],
                [0, 6 * np.pi]])

L_p = np.array([[-3 * np.pi / 17, 0],
                [0, -6 * np.pi]])

L = np.block([[np.zeros((2, 2)), L_e],
              [L_p, np.zeros((2, 2))]])
```

```python
def dsdt(t, s, U):
    return W @ s + U * (L @ s)

def dUdt(t, s, U):
    psi_e, _, psi_p, _ = s
    return 0.25 * k * (k_p_squared - k_e_squared) * psi_e * psi_p - r_m * U
```

```python
def rhs(t, y):
    s, U = y[:4], y[4]
    ds = dsdt(t, s, U)
    dU = dUdt(t, s, U)
    return np.concatenate((ds, [dU]))

initial_state = np.array([0.1, 0.1, 0, 0, 0.001])
t_span = (0, 6)

sol = solve_ivp(rhs, t_span, initial_state, method='RK45', t_eval=np.linspace(t_span[0], t_span[1], 10000), vectorized=True, dense_output=True)
```

```python
t = np.linspace(t_span[0], t_span[1], 10000)

y = sol.sol(t)
R_value = 0.25 * k * (k_p_squared - k_e_squared) * y[0] * y[2]

fig, axs = plt.subplots(2, 1, figsize=(10, 8))
axs[0].plot(t, R_value,'-')
axs[0].set_title("Time evolution of R")
axs[0].set_ylabel("R")

axs[1].plot(t, y[4],'-')
axs[1].set_title("Time evolution of U")
axs[1].set_xlabel("Time")
axs[1].set_ylabel("U")

plt.tight_layout()
plt.show()
```
